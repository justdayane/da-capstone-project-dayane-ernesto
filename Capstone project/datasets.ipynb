```python
df11 = pd.read_csv("./consumption_levels/data/SA_0000001823.csv")
df11
```

```python
df_analysis = df11[[
    'SpatialDimensionValueCode',  # country code
    'ParentLocation',             # region (optional)
    'TimeDim',                    # year
    'Value',
    'NumericValue',
    'Low',
    'High'
]]
df11.head()
```

```python
df_analysis = df_analysis.rename(columns={
    'SpatialDimensionValueCode': 'country',
    'TimeDim': 'year'
})
```

```python
df_analysis['NumericValue'] = pd.to_numeric(df_analysis['NumericValue'], errors='coerce')
df_analysis['Low'] = pd.to_numeric(df_analysis['Low'], errors='coerce')
df_analysis['High'] = pd.to_numeric(df_analysis['High'], errors='coerce')
```

```python
df_grouped = df_analysis.groupby(['country', 'year']).agg({
    'NumericValue': 'mean',
    'Low': 'mean',
    'High': 'mean'
}).reset_index()
```

```python
import plotly.express as px

df_analysis = df11[['SpatialDimensionValueCode', 'TimeDim', 'NumericValue']].copy()
df_analysis.columns = ['country', 'year', 'value']
```

```python
latest_year = df_analysis['year'].max()

df_year = df_analysis[df_analysis['year'] == latest_year]

fig = px.choropleth(
    df_year,
    locations="country",          # ISO-3 codes
    color="value",
    hover_name="country",
    color_continuous_scale="Viridis",
    title=f"Consumption by Country ({latest_year})"
)

fig.show()
```

```python
import pycountry

def get_country_name(code):
    try:
        return pycountry.countries.get(alpha_3=code).name
    except:
        return code

df_analysis['country_name'] = df_analysis['country'].apply(get_country_name)
```

```python
hover_name="country_name"
```

```python
fig = px.choropleth(
    df_analysis,
    locations="country",
    color="value",
    hover_name="country",
    animation_frame="year",
    color_continuous_scale="RdYlBu",
    range_color=(-0.5, 0.5),
    title="Consumption by Country Over Time"
)

fig.show()
```
