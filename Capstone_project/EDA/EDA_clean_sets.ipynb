```python
df1 = pd.read_csv("/Users/dayaneribeiro/Documents/Spiced/da-capstone-project-dayane-ernesto/Cleaned Sets/consumption_and_prevalence_2005-2020.csv")
df1
```

```python
df1["precent_of_prevalence"] = df1["precent_of_prevalence"] * 100
df1.head()
```

```python
fig = px.choropleth(
    df1,
    #df1_clean.sort_values(by="NumericValue", ascending=False),
    locations="location", 
    locationmode="country names",        
    color="precent_of_prevalence",
    hover_name="location",
    #hover_data="year",
    #hover_data="age",
    color_continuous_scale="RdYlBu",
    #range_color=(0, 1),
    title="Alcohol Consumption by Country"
)

fig.show()
```

```python
df_age = df1[df1["age"] == "15-19 years"]

fig = px.choropleth(
    df_age,
    locations="location",
    locationmode="country names",
    color="precent_of_prevalence",
    hover_name="location",
    color_continuous_scale="RdYlBu",
    title="People with AUD by Country / Age"
)

fig.show()
```
