```python
df2 = pd.read_csv("/Users/dayaneribeiro/Documents/Spiced/da-capstone-project-dayane-ernesto/Cleaned Sets/master_prevalence_2005_2023.csv")
df2
```

```python
df2["precent_of_prevalence"] = df2["precent_of_prevalence"] * 100
df2.head()
```

```python
df2.rename(columns={"precent_of_prevalence": "AUD_percentage", "location": "country"})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_plot = df2[
    (df2["year"] >= 2010) & 
    (df2["year"] <= 2020) &
    (df2["sex"] == "Both Sexes")
]

plt.figure(figsize=(10, 6))

sns.barplot(
    data=df_plot,
    x="year",
    y="precent_of_prevalence",
    hue="age"  
)

plt.title("AUD Percentage by Age Group Over Time")
plt.xlabel("Year")
plt.ylabel("AUD Percentage")

plt.legend(title="Age Group", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.tight_layout()
plt.show()
```

```python
countries = ["Germany", "France", "Poland", "Czechia", "Ireland"]

df_subset = df3[df3["Title"].isin(countries)]

plt.figure(figsize=(10,6))
sns.lineplot(
    data=df_subset,
    x="Year",
    y="NumericValue",
    hue="Title"
)

plt.title("Alcohol Consumption Over Time")
plt.show()
```

```python
fig = px.line(
    df_subset,
    x="Year",
    y="NumericValue",
    color="Title"
)

fig.update_layout(
    updatemenus=[
        dict(
            buttons=[
                dict(
                    label=country,
                    method="update",
                    args=[{"visible": [c == country for c in df3["Title"].unique()]}]
                )
                for country in df3["Title"].unique()
            ],
            direction="down"
        )
    ]
)

fig.show()
```
