```python
gen_analysis = pd.read_csv("/Users/dayaneribeiro/Documents/Spiced/da-capstone-project-dayane-ernesto/Cleaned Sets/gen_gap_2019-analysis.csv")
gen_analysis
```

```python
gen_analysis.rename(columns={"precent_of_prevalence": "Prevalence", "location": "Country"}, inplace=True)
```

```python
gen_analysis["prevalence"] = gen_analysis["prevalence"] * 100
gen_analysis.head()
```

```python

def define_gen(age):
    if '15-19 years' in age or '20-24 years' in age:
        return 'Gen_Z'
    elif '25-29 years' in age or '30-34 years' in age or '35-39 years' in age:
        return 'Millennial'
    else:
        return 'Gen_X'
```

```python
gen_analysis['generation'] = gen_analysis['age'].apply(define_gen)
gen_analysis.head()
```

```python
# 3. (Optional but recommended) filter to avoid duplicates across sex
df_filtered = gen_analysis[gen_analysis['sex'] == 'Both Sexes']

# 4. Aggregate mean prevalence by year and generation
grouped = df_filtered.groupby(['year', 'generation'])['prevalence'].mean().reset_index()

# 5. Pivot for plotting
pivot_df = grouped.pivot(index='year', columns='generation', values='prevalence')
```

```python
fig = px.choropleth(
    gen_analysis,
    locations="location",
    locationmode="country names",
    color="prevalence",
    hover_name="location",
    animation_frame="generation", 
    color_continuous_scale="RdYlBu",
    title="AUD prevalence in Generations"
)
fig.update_geos(
    #scope="europe",
    projection_scale=2.5,
    center={"lat": 54, "lon": 15} 
)
#fig.layout.sliders[0].transition.duration = 2000
fig.show()
```
