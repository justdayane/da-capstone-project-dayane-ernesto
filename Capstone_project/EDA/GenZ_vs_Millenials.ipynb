```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import plotly as plt
```

```python
df = pd.read_csv("/Users/dayaneribeiro/Documents/Spiced/da-capstone-project-dayane-ernesto/Cleaned Sets/gen_gap_2019-analysis.csv")
df
```

```python
df.info()
```

```python
df = df.rename(columns = {'precent_of_prevalence': 'prevalence'}, inplace=True)
```

```python

def define_gen(age):
    if '15-19 years' in age or '20-24 years' in age:
        return 'Gen_Z'
    elif '25-29 years' in age or '30-34 years' in age or '35-39 years' in age:
        return 'Millennial'
    else:
        return 'Gen_X'
```

```python
fig = px.choropleth(
    df,
    locations="location",
    locationmode="country names",
    color="prevalence",
    hover_name="location",
    animation_frame="age", 
    color_continuous_scale="RdYlBu",
    title="AUD prevalence by Country and Age"
)
fig.update_geos(
    #scope="europe",
    projection_scale=2.5,
    center={"lat": 54, "lon": 15} 
)
#fig.layout.sliders[0].transition.duration = 2000
fig.show()
```

```python
df['generation'] = df['age'].apply(define_gen)
df.head()
```

```python

import matplotlib.pyplot as plt

# 3. (Optional but recommended) filter to avoid duplicates across sex
df_filtered = df[df['sex'] == 'Both Sexes']

# 4. Aggregate mean prevalence by year and generation
grouped = df_filtered.groupby(['year', 'generation'])['precent_of_prevalence'].mean().reset_index()

# 5. Pivot for plotting
pivot_df = grouped.pivot(index='year', columns='generation', values='precent_of_prevalence')

# 6. Plot
plt.figure(figsize=(10,6))
sns.lineplot(
    data=pivot_df
)

plt.xlabel('Year')
plt.ylabel('Prevalence')
plt.title('Prevalence by Generation Over Time')
plt.legend(title="Generation", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.show()
```
