```python
countries = ["Germany", "Spain", "France", "Italy"]

df1_subset = df1[df1["Title"].isin(countries)]

plt.figure(figsize=(12,6))
sns.barplot(
    data=df1_subset,
    x="Year",
    y="NumericValue",
    hue="Title"
)

plt.xticks(rotation=45)
plt.title("Alcohol Consumption per year")
plt.show()
```

```python

fig = px.line(
    df1,
    x="Year",
    y="NumericValue",
    color="Title",
    title="Alcohol Consumption Over Time by Country",
    labels={"NumericValue": "Liters per capita"}
)

fig.show()
```

```python
countries = ["Germany", "Spain", "France", "Italy"]

df1_subset = df1[df1["Title"].isin(countries)]

fig = px.line(
    df1_subset,
    x="Year",
    y="Liters Pure Alcohol Consumed",
    color="Title",
    markers=True,
    title="Alcohol Consumption (Selected Countries)"
)

fig.show()
```

```python
fig = px.line(
    df1,
    x="Year",
    y="NumericValue",
    color="Title"
)

fig.update_layout(
    updatemenus=[
        dict(
            buttons=[
                dict(
                    label=country,
                    method="update",
                    args=[{"visible": [c == country for c in df1["Title"].unique()]}]
                )
                for country in df1["Title"].unique()
            ],
            direction="down"
        )
    ]
)

fig.show()
```
