```python

fig = px.line(
    df1,
    x="Year",
    y="NumericValue",
    color="Title",
    title="Alcohol Consumption Over Time by Country",
    labels={"NumericValue": "Liters per capita"}
)

fig.show()
```

```python
countries = ["Germany", "Spain", "France", "Italy"]

df1_subset = df1[df1["Title"].isin(countries)]

fig = px.line(
    df1_subset,
    x="Year",
    y="Liters Pure Alcohol Consumed",
    color="Title",
    markers=True,
    title="Alcohol Consumption (Selected Countries)"
)

fig.show()
```

```python
fig = px.line(
    df1,
    x="Year",
    y="NumericValue",
    color="Title"
)

fig.update_layout(
    updatemenus=[
        dict(
            buttons=[
                dict(
                    label=country,
                    method="update",
                    args=[{"visible": [c == country for c in df1["Title"].unique()]}]
                )
                for country in df1["Title"].unique()
            ],
            direction="down"
        )
    ]
)

fig.show()
```

```python
fig = px.choropleth(
    df1,
    locations="Title",
    color="NumericValue",
    hover_name="Title",
    animation_frame="Year",
    color_continuous_scale="RdYlBu",
    range_color=(-1, 2),
    title="Consumption by Country Over Time"
)

fig.show()
```
