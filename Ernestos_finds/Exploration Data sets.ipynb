```python
cluster_df = pd.read_csv('/Users/nestocr/Documents/DataAnalytics_BootCamp/da-capstone-project-dayane-ernesto/Cleaned Sets/clustering_ready_2019.csv')
```

```python
cluster_df.rename(columns={'val':'prevalence','liters':'liters/capita'}, inplace=True)
```

```python
cluster_df.head(5)
```

```python
scaler = StandardScaler()
```

```python
features = ['prevalence','liters/capita','abstention_rate']
X=cluster_df[features]
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
kmeans=KMeans(n_clusters=4, random_state=42, n_init=10)
cluster_df['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
cluster_names = {
    0: 'Mainstream Drinking Cultures',
    1: 'High Consumption Cultures',
    2: 'Abstainer-Dominant Cultures',
    3: 'High-Risk Drinking Cultures'
}

cluster_df['cluster'] = cluster_df['cluster'].map(cluster_names)
```

```python
cluster_df['cluster'] = cluster_df['cluster'].astype(str)
```

```python
cluster_df.groupby('cluster')[['liters/capita', 'abstention_rate', 'prevalence']].mean()
```

```python
fig = px.scatter_3d(cluster_df, 
                    x='liters/capita', 
                    y='abstention_rate', 
                    z='prevalence',
                    color='cluster',
                    hover_name='location',
                    title='The 4 Drinking Cultures of Europe (2019)',
                    template='plotly_dark',
                    width=1000,
                    height=1000,    
                    labels={'val': 'Prevalence (Disorders)', 
                            'liters': 'Liters per Capita', 
                            'abstention_rate': '% Abstainers'})

fig.show()
```

```python
fig.write_image('clusters.png', width =1000, height=1000, scale=2)
```

```python
fig_bubble = px.scatter(cluster_df,
                        x='liters/capita',
                        y='abstention_rate',
                        size='prevalence',
                        color='cluster',
                        hover_name='location',
                        title='The 4 Drinking Cultures of Europe (2019)',
                        template='plotly_dark',
                        width=1000,
                        height=1000,
                        size_max=40,                # controls max bubble size
                        labels={'liters/capita': 'Liters per Capita',
                                'abstention_rate': '% Abstainers',
                                'prevalence': 'Prevalence (Disorders)'})

fig_bubble.show()
```

```python
# Parallel coordinates needs a numeric color value, so we map the cluster labels
cluster_map = {label: i for i, label in enumerate(cluster_df['cluster'].unique())}
cluster_df['cluster_num'] = cluster_df['cluster'].map(cluster_map)

fig_parallel = px.parallel_coordinates(cluster_df,
                                        color='cluster_num',
                                        dimensions=['liters/capita', 'abstention_rate', 'prevalence'],
                                        title='The 4 Drinking Cultures of Europe (2019)',
                                        template='plotly_dark',
                                        width=1000,
                                        height=700,
                                        color_continuous_scale=px.colors.qualitative.Safe,
                                        labels={'liters/capita': 'Liters per Capita',
                                                'abstention_rate': '% Abstainers',
                                                'prevalence': 'Prevalence (Disorders)',
                                                'cluster_num': 'Cluster'})

# Add the original cluster names as colorbar tick labels
fig_parallel.update_layout(
    coloraxis_colorbar=dict(
        tickvals=list(cluster_map.values()),
        ticktext=list(cluster_map.keys())
    )
)

fig_parallel.show()
```

```python
sns.set_theme(style="whitegrid")
g = sns.pairplot(cluster_df,
                 hue='cluster',
                 palette='CMRmap',
                 vars=['prevalence', 'liters/capita', 'abstention_rate'],
                 diag_kind='kde',
                 height=3)
g.fig.suptitle('Drinking Culture Clusters: 2D Relationship Matrix (2019)',y=1.02, fontsize=16)
py.show()
```

```python
K = range(2, 10) #let's give it a range
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234, # a different random state 
                    n_init=10)
    kmeans.fit(cluster_df[features])
    inertia.append(kmeans.inertia_)
```

```python
py.figure(figsize=(16,8))
py.plot(K, inertia, 'bx-') # shows the x symbols on the graph
py.xlabel('k')
py.ylabel('inertia')
py.xticks(np.arange(min(K), max(K), 1.0))
py.title('Elbow Method showing the optimal k')
```

```python
def define_gen(row):
    age_lower = int(row['age'].split('-')[0])
    birth_year = row['year'] - age_lower
    if birth_year >= 1997:
        return 'Gen_Z'
    elif birth_year >= 1981:
        return 'Millennial'
    else: 
        return'Gen_X'

gen_gap['generation'] = gen_gap.apply(define_gen, axis=1)
```

```python
def assign_generation_2019(age):
    if age == '15-19 years':
        return 'Gen Z'
    elif age in ['20-24 years', '25-29 years', '30-34 years', '35-39 years']:
        return 'Millennial'
    elif age == '40-44 years':
        return 'Gen X'
    else:
        return 'Other'
    
snapshot_2019 = master_prevalence[
        (master_prevalence['year'] == 2019) &
        (master_prevalence['sex'] == 'Both Sexes')
].copy()

snapshot_2019['generation'] = snapshot_2019['age'].apply(assign_generation_2019)

# Verify the output
print(snapshot_2019.groupby(['generation', 'age'])['precent_of_prevalence'].mean().round(4))


```

```python
# Filter for the peer comparison: same age cohort, two different years
# Using 20-24 as the focal cohort since it's the most discussed

peer_df = master_prevalence[
    (master_prevalence['age'] == '20-24 years') &
    (master_prevalence['sex'] == 'Both Sexes') &
    (master_prevalence['year'].isin([2010, 2019]))
].copy()

# Pivot so 2010 and 2019 are side by side per country
peer_pivot = peer_df.pivot_table(
    index='location',
    columns='year',
    values='precent_of_prevalence'
).round(4)

# Calculate relative percentage change
peer_pivot['change_pct'] = (
    (peer_pivot[2019] - peer_pivot[2010]) / peer_pivot[2010] * 100
).round(1)

peer_pivot.columns = ['Prevalence_2010', 'Prevalence_2019', 'Relative_Change_%']

print(peer_pivot.sort_values('Relative_Change_%'))
```
