```python
cluster_df = pd.read_csv('/Users/nestocr/Documents/DataAnalytics_BootCamp/da-capstone-project-dayane-ernesto/Cleaned Sets/clustering_ready_2019.csv')
```

```python
cluster_df.rename(columns={'val':'prevalence','liters':'liters/capita'}, inplace=True)
```

```python
scaler = StandardScaler()
```

```python
features = ['prevalence','liters/capita','abstention_rate']
X=cluster_df[features]
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
kmeans=KMeans(n_clusters=4, random_state=42, n_init=10)
```

```python
cluster_df['cluster'] =kmeans.fit_predict(X_scaled)
```

```python
cluster_df['cluster'] = cluster_df['cluster'].astype(str)
```

```python
fig = px.scatter_3d(cluster_df, 
                    x='liters/capita', 
                    y='abstention_rate', 
                    z='prevalence',
                    color='cluster',
                    hover_name='location',
                    title='The 4 Drinking Cultures of Europe (2019)',
                    template='plotly_dark',
                    width=1000,
                    height=1000,    
                    labels={'val': 'Prevalence (Disorders)', 
                            'liters': 'Liters per Capita', 
                            'abstention_rate': '% Abstainers'})

fig.show()
```

```python
sns.set_theme(style="whitegrid")
g = sns.pairplot(cluster_df,
                 hue='cluster',
                 palette='CMRmap',
                 vars=['prevalence', 'liters/capita', 'abstention_rate'],
                 diag_kind='kde',
                 height=3)
g.fig.suptitle('Drinking Culture Clusters: 2D Relationship Matrix (2019)',y=1.02, fontsize=16)
py.show()
```

```python
K = range(2, 10) #let's give it a range
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234, # a different random state 
                    n_init=10)
    kmeans.fit(cluster_df[features])
    inertia.append(kmeans.inertia_)
```

```python
py.figure(figsize=(16,8))
py.plot(K, inertia, 'bx-') # shows the x symbols on the graph
py.xlabel('k')
py.ylabel('inertia')
py.xticks(np.arange(min(K), max(K), 1.0))
py.title('Elbow Method showing the optimal k')
```

```python
def define_gen(row):
    age_lower = int(row['age'].split('-')[0])
    birth_year = row['year'] - age_lower
    if birth_year >= 1997:
        return 'Gen_Z'
    elif birth_year >= 1981:
        return 'Millennial'
    else: 
        return'Gen_X'

gen_gap['generation'] = gen_gap.apply(define_gen, axis=1)
```

```python
peer_df = gen_gap[
            (gen_gap['age']=='20-24 years') & 
             (gen_gap['year'].isin([2010, 2019])) & 
             (gen_gap['sex'] == 'Both Sexes')].copy()
```

```python
peer_df
```

```python
pivot_peer = peer_df.pivot(index='location', columns='year', values='precent_of_prevalence')
pivot_peer['Absolute Change'] = pivot_peer[2019] - pivot_peer[2010]
pivot_peer['% Relative Change'] = (pivot_peer['Absolute Change'] / pivot_peer[2010]) * 100
```

```python
pivot_peer.head()
```

```python
py.figure(figsize=(10,7))
sns.set_style('white')

for i in pivot_peer.index: 
    py.plot([2010,2019], [pivot_peer.loc[i, 2010],pivot_peer.loc[i, 2019]], marker='o', label = i, linewidth = 2)
    py.text(2009.8,pivot_peer.loc[i,2010], f'{i}:{pivot_peer.loc[i, 2010]*100:.2f}%',horizontalalignment='right', fontsize=10)
    py.text(2019.2,pivot_peer.loc[i,2019], f'{i}:{pivot_peer.loc[i, 2019]*100:.2f}%',horizontalalignment='left', fontsize=10)

py.title("The '20-Year-Old' Risk Profile:2010 (Millennial) vs. 2019 (Gen Z)",fontsize=14, pad = 20)
py.ylabel("Prevalence of Alcohol DIsorders (%)")
py.xticks([2010,2019], ["Millennial (in 2010)", "Gen Z (in 2019)"])
py.xlim(2008, 2021)
py.grid(axis='y', linestyle='--', alpha=0.3)
py.tight_layout()
py.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:.2%}'))


py.savefig('peer_comparison_slope_chart.png')

py.show()

```
